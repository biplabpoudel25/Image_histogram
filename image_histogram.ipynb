```python
# Import libraries
import cv2
import time
import matplotlib.pyplot as plt
```

```python
# define the user defined function
def compute_histogram(image, nbins):
    freq = [ (image == px).sum() for px in range(nbins)]
    return freq
```

```python
nbins = 256
img_path = r'lenna_img.png'
img = cv2.imread(img_path, 0)
```

```python
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.show()
```

```python
tic = time.time()
freq = compute_histogram(img, nbins)
toc = time.time()
print("Total time elapsed: " + str(1000*(toc - tic)) + ' ms')
```

```python
plt.bar(range(nbins), freq)
plt.xlabel("Intensity values")
plt.ylabel("frequency")
plt.title(" Histogram of original image")
plt.show()
```
